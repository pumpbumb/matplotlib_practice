```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread(r'D:\code\数据分析\matplotlib_practice\image_tutorial\image\stinkbug.png')
print(img)
```

```python
# Plotting numpy arrays as images
imgplot = plt.imshow(img)
```

```python
a = np.diag(range(10))  # (M, N): an image with scalar data.                  
# print(a)                                 
imgplot_a = plt.imshow(a)
```

```python
# 随机生成 [0, 1) 之间的浮点数    Return random floats in the half-open interval [0.0, 1.0). 
print('一维数组:\n', np.random.random(3), '\n')
print('二维数组:\n', np.random.random((2, 3)), '\n')
print('三维数组:\n', np.random.random((3, 2, 4)))
```

```python
#  生成具有某种分布的随机数
# Call default_rng to get a new instance of a `Generator`, then call its methods to obtain samples from different distributions. 
from numpy.random import default_rng  # rng  是指 random generator 
rng = default_rng()  
print(rng, '\n')

vals = rng.standard_normal(10)  # 这里是标准正态分布
print(vals, '\n')
```

```python
x1 = np.random.random((100, 100, 3))  # # (100, 100, 3): an image with RGB values (0-1 float or 0-255 int).
imgplot_x1 = plt.imshow(x1)
```

# 将伪彩色方案应用于image plots

```python
lum_img = img[:, :, 0]  # 提取R通道
print(type(lum_img), '\n', lum_img.shape)  # 2D array
lum_img_imshow = plt.imshow(lum_img, cmap='hot', clim=(0.0, 0.9))
plt.colorbar()

# Note that you can also change colormaps on existing plot objects using the set_cmap() method:
# lum_img_imshow.set_cmap('nipy_spectral')  
```

```python
# 对以上切片作解释
lum_image_explain = np.arange(12, 255, 9).reshape(3, 3, 3)
print(lum_image_explain)
plt.imshow(lum_image_explain)
plt.colorbar()
```

```python
print(lum_image_explain[:, :, 0])  # 切片
plt.imshow(lum_image_explain[:, :, 0], cmap='hot')
plt.colorbar(orientation='horizontal')  # 颜色条设置为水平放置
```
